# Questão 1

Para um plano de amostragem simples com amostras de tamanho 100 e aceitação para 2 itens defeituosos determinar o NQA para um risco α igual a 7%.

```python
def NQASimples(Ac, n, alpha, precisao = 0.000001): #Criando uma rotina para determinar o valor de NQA para amostragem simples.
  n = abs(int(n)) #Tamanho da amostra.
  Ac = abs(int(Ac)) #Número máximo de defeituosos para aceitação do lote.
  alpha = abs(float(alpha)) #Risco do produtor.
  NQA = 0 #Nível de Qualidade Aceitável.
  precisao = abs(float(precisao))
  
  while not alpha <= 1 - binom.cdf(k = Ac, n = n, p = NQA):
    NQA += precisao

  return NQA

NQASimples(n = 100, Ac = 2, alpha = 0.07) # = 0.947 %
```

# Questão 2

Um plano de amostragem duplo é definido por amostras de tamanho 50 e aceitação para 2 itens defeituosos e rejeição para 5 defeituosos na primeira amostra e amostras de tamanho 200 na segunda amostra. Determinar o NQA correspondente a um risco α igual a 7%.

```python
def NQADupla(Ac, Re, n1, n2, alpha, precisao = 0.000001): #Criando uma rotina para determinar o valor de NQA para amostragem dupla.
  n1 = abs(int(n1)) #Tamanho da primeira amostra.
  n2 = abs(int(n2)) #Tamanho da segunda amostra.
  Ac = abs(int(Ac)) #Número máximo de defeituosos na primeira amostra para a aceitação do lote.
  Re = abs(int(Ac)) #Número máximo de defeituosos em ambas as amostras para a aceitação do lote.
  alpha = abs(float(alpha)) #Risco do produtor.
  NQA = 0 #Nível de Qualidade Aceitável.
  precisao = abs(float(precisao)) #Estabelece o incremento para NQA. Quanto mais próximo de zero, mais preciso será o NQA.
  P = 1 #Probabilidade de se aceitar o lote no nível de qualidade aceitável.
  
  while not alpha <= 1 - P:
    P1 = binom.cdf(k = Ac, n = n1, p = NQA) #Probabilidade de aceitar o lote na primeira amostra.
    P2 = 0 #Probabilidade de se aceitar o lote na segunda amostra.
    i = 1 #Define-se um contador que será usado na estrutura de iteração a seguir.
    while i <= Re - Ac:
      P2 += binom.pmf(k = Ac + i, n = n1, p = NQA) * binom.cdf(k = Re - Ac - i, n = n2, p = NQA)
      i += 1
    P = P1 + P2 # Aprobabilidade de se acietar o lote será a combinação conjunta de P1 e P2.
    NQA += precisao #...NQA vai sendo incrementado até atingir um valor que satisfaça a condição.
  return NQA #Por fim, retorna NQA.

NQADupla(Ac = 2, Re = 5, n1 = 50, n2 = 200, alpha = 0.07) # = 1.905 %
```

# Questão 3

Determinar o tamanho mínimo da amostra e o correspondente número de aceitação para os seguintes valores: NQA = 7,0%, NQI = 15,0%, $\alpha$ = 5,0% e $\beta$ = 10,0%

```python
def NMinimo(NQA, NQI, alpha, beta): #Função para determinar o valor mínimo da amostra.
  NQA = abs(float(NQA)) #Nível de Qaulidade Aceitável.
  NQI = abs(float(NQI)) #Nível de Qualidade Inaceitável.
  alpha = abs(float(alpha)) #Risco do produtor.
  beta = abs(float(beta)) #Risco do consumidor.
  n = 1 #Tamanho da amostra a ser calculada.
  Ac = 0 #Número de aceitação a ser calculada.

  while binom.cdf(Ac, n, NQI) > beta or 1 - binom.cdf(Ac, n, NQA) > alpha:
    if binom.cdf(Ac, n, NQI) > beta:
      n += 1
      """O aumento de n acarreta na diminuição de binom.cdf(Ac, n, NQI) para um dado valor de Ac.
      E também o aumento de Ac acarreta no aumento deste cdf. Logo, se a condição acima é satisfeita,
      então não adianta aumenta Ac pois não encontraremos a solução, só restando aumentar n para
      diminuir o cdf de beta."""
      Ac = 0
      """Na verdade, penso que a reinicialização da variável Ac não seria necessária, pois o incremento de Ac
      aumenta o cdf de beta, enquanto diminui o cdf negativo de alpha. Logo, se para um dado Ac os
      cdf's não não menores do que seus respectivos riscos para um dado Ac0, certamente eles não seriam
      menores para valores menores que Ac0. No entanto, como não posso assegurar que isso vá ocorrer para
      quaisquer valores reais de NQA, NQI, alpha e beta, seria melhor reiniciar a variável Ac."""
    elif 1 - binom.cdf(Ac, n, NQA) > alpha:
      Ac += 1

  return {"n": n, "Ac": Ac} # Retorna um dicionário com os valores de n e Ac.

NMinimo(NQA = 0.07, NQI = 0.15, alpha = 0.05, beta = 0.10) # Retorna {'Ac': 14, 'n': 132}
```
